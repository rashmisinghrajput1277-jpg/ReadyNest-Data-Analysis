```python
# ==========================================
# STEP 1: LOAD THE LIBRARIES WE NEED



# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Step 1 Complete: Libraries loaded smoothly!")

# ==========================================
# STEP 2: CREATE THE DATASET FROM SCRATCH
# ==========================================

np.random.seed(42)
total_records = 200

mock_data = {
    'Property_ID': [f"RN-{1000+i}" for i in range(total_records)],
    'Region': np.random.choice(['North', 'South', 'East', 'West'], size=total_records, p=[0.3, 0.25, 0.2, 0.25]),
    'Property_Type': np.random.choice(['Studio Capsule', 'Family Modular', 'Duplex Prefab', 'Eco-Pod'], size=total_records),
    'Construction_Cost_INR': np.random.randint(1500000, 4500000, size=total_records),
    'Sale_Price_INR': np.random.randint(2000000, 6500000, size=total_records),
    'Units_Sold': np.random.randint(1, 12, size=total_records),
    'Customer_Rating': np.random.choice([1, 2, 3, 4, 5, np.nan], size=total_records, p=[0.05, 0.05, 0.1, 0.4, 0.35, 0.05])
}

# Convert this dictionary into a spreadsheet format called a DataFrame
df = pd.DataFrame(mock_data)
print("\nStep 2 Complete: Mock dataset created successfully!")

# ==========================================
# STEP 3: CLEAN THE DATA (The "Broom & Wrench" Phase)
# ==========================================
# 1. Fill any missing customer ratings with the average (median) rating
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())

# 2. Add a calculated column for total business profit
df['Total_Profit_INR'] = (df['Sale_Price_INR'] - df['Construction_Cost_INR']) * df['Units_Sold']

print("\nStep 3 Complete: Data has been cleaned and processed.")

# ==========================================
# STEP 4: CALCULATE THE STATISTICS (Sigma Rule)
# ==========================================
print("\n--- STATISTICAL SUMMARY (SIGMA) ---")
print(df[['Construction_Cost_INR', 'Sale_Price_INR', 'Total_Profit_INR']].describe())

# ==========================================
# STEP 5: VISUALIZE THE DATA (Charts)
# ==========================================
plt.figure(figsize=(12, 5))

# Chart 1: Bar chart showing average profit for each type of home
plt.subplot(1, 2, 1)
sns.barplot(x='Property_Type', y='Total_Profit_INR', data=df, errorbar=None, palette='pastel')
plt.title('Average Profit by Property Type')
plt.xticks(rotation=15)

# Chart 2: Scatter plot showing relation between Cost and Sale Price
plt.subplot(1, 2, 2)
sns.scatterplot(x='Construction_Cost_INR', y='Sale_Price_INR', hue='Region', data=df)
plt.title('Construction Cost vs Sale Price')

plt.tight_layout()
plt.show()

# ==========================================
# STEP 6: SAVE THE FILE
# ==========================================
df.to_csv('cleaned_readynest_data.csv', index=False)
print("\nStep 6 Complete: Saved clean data as 'cleaned_readynest_data.csv'")
```
